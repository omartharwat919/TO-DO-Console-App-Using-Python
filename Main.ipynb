# Main Menu

```python
class MainMenu:
    def __init__(self):
        self.logged_in_email = None
        self.display_menu()
    def display_menu(self):
        
        while True:
            print("\n📌 Main Menu")
            print("1. Authentication System")
            print("2. Task Management System")
            print("3. Exit")

            try:
                choice = input("Choose an option: ").strip()

                if choice == '1':
                    auth = AuthenticationSystem()
                    self.logged_in_email = auth.logged_in_user.email  

                elif choice == '2':
                    if not self.logged_in_email:  
                        print("❌ You need to log in first!")
                        auth = AuthenticationSystem()
                        self.logged_in_email = auth.logged_in_user.email  

                    
                    task_system = TaskManagementSystem(self.logged_in_email)
                    
                    
                elif choice == '3':
                    print("👋 Goodbye!")
                    break  

                else:
                    print("❌ Invalid choice. Please try again.")

            except KeyboardInterrupt:
                print("\n❌ Program interrupted by user. Exiting...")
                return  # Exit safely when Ctrl+C is pressed


```
